```python
import matplotlib.pyplot as plt
import numpy as np

# Simple data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot
plt.figure(figsize=(6,4))
plt.plot(x, y, label="sin(x)")
plt.title("Example Plot in about.ipynb")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```
