## Python Demo

The following Python code cell is folded by default, but the output plot is shown.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 200)
y = np.cos(x)

plt.figure(figsize=(6,4))
plt.plot(x, y, color="darkorange", label="cos(x)")
plt.title("Cosine Function Example")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.legend()
plt.grid(True)
plt.show()
```
